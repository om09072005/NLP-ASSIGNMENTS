```python
import spacy
from seqeval.metrics import classification_report

# Load the English pipeline
nlp = spacy.load("en_core_web_sm")
```

```python
def extract_entities(text):
    doc = nlp(text)
    print(f"{'Text':<15} | {'Label'}")
    print("-" * 25)
    for ent in doc.ents:
        print(f"{ent.text:<15} | {ent.label_}")
    return doc
text_data = "Google is headquartered in Mountain View and was founded by Larry Page."
doc_output = extract_entities(text_data)
```

```python
y_true = [['B-ORG', 'O', 'O', 'O', 'B-GPE', 'I-GPE']]
y_pred = [['B-ORG', 'O', 'O', 'O', 'B-GPE', 'O']]
print("\nAccuracy Metrics Report:")
print(classification_report(y_true, y_pred))
```
