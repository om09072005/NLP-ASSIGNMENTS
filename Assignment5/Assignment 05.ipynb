```python
import nltk
from nltk.corpus import wordnet as wn

nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

print("WordNet and NLTK data loaded successfully.")
```

```python
def get_semantic_info(word):
    synsets = wn.synsets(word)

    if not synsets:
        return f"No synsets found for '{word}'"

    print(f"--- Analysis for '{word}' ---")
    for i, synset in enumerate(synsets[:3]):
        print(f"\nSense {i+1}: {synset.name()}")
        print(f"Definition: {synset.definition()}")

        synonyms = [lemma.name().replace('_', ' ') for lemma in synset.lemmas()]
        print(f"Synonyms: {set(synonyms)}")
get_semantic_info("happy")
```

```python
def get_antonyms(word):
    antonyms = []
    for synset in wn.synsets(word):
        for lemma in synset.lemmas():
            if lemma.antonyms():
                antonyms.append(lemma.antonyms()[0].name())

    print(f"Antonyms for '{word}': {set(antonyms)}")


get_antonyms("wise")
```

```python
def get_hierarchy(word):
    synsets = wn.synsets(word)
    if not synsets: return

    primary_synset = synsets[0]
    hyper = primary_synset.hypernyms()
    print(f"Hypernyms (Parent Category) of {word}: {[h.name() for h in hyper]}")
    hypo = primary_synset.hyponyms()
    print(f"Hyponyms (Specific Types) of {word}: {[h.name() for h in hypo[:5]]}")
get_hierarchy("dog")
```
