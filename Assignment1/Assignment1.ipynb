```python
import nltk
from nltk.tokenize import (
    word_tokenize,
    WhitespaceTokenizer,
    TreebankWordTokenizer,
    TweetTokenizer,
    MWETokenizer
)
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

Whitespace Tokenization

```python
whitespace_tokenizer = WhitespaceTokenizer()
tokens_ws = whitespace_tokenizer.tokenize(text)
print("Whitespace Tokenization:", tokens_ws)
```

Punctuation-based Tokenization

```python
tokens_punct = word_tokenize(text)
print("Punctuation-based Tokenization:", tokens_punct)
```

Treebank Tokenization

```python
treebank_tokenizer = TreebankWordTokenizer()
tokens_treebank = treebank_tokenizer.tokenize(text)
print("Treebank Tokenization:", tokens_treebank)
```

Multi-Word Expression Tokenization

```python
mwe_tokenizer = MWETokenizer([('machine', 'learning'), ('natural', 'language', 'processing')])
mwe_text = "I am studying machine learning and natural language processing"
tokens_mwe = mwe_tokenizer.tokenize(mwe_text.split())
print("MWE Tokenization:", tokens_mwe)
```

#STEMMING
Porter Stemmer

```python
porter = PorterStemmer()
words = ["running", "studies", "better", "connection"]

porter_stems = [porter.stem(word) for word in words]
print("Porter Stemmer:", porter_stems)
```

Snowball Stemmer

```python
snowball = SnowballStemmer("english")
snowball_stems = [snowball.stem(word) for word in words]
print("Snowball Stemmer:", snowball_stems)
```

#LEMMATIZATION
Lemmatization using WordNet

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = [
    lemmatizer.lemmatize("running", pos='v'),
    lemmatizer.lemmatize("studies", pos='n'),
    lemmatizer.lemmatize("better", pos='a')
]

print("Lemmatization:", lemmatized_words)
```
